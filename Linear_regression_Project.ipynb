```python
def outliers(df):
  q1 = df.quantile(0.25)
  q3 = df.quantile(0.75)
  IQR = q3 -q1
  lower = q1 - 1.5 * IQR
  upper = q3 + 1.5 * IQR
  outliers = df[(df < lower) | (df > upper)]
  return outliers

column_names = df.select_dtypes(include = ['float64', 'int64'])

for i in column_names:
  result = outliers(df[i])
  print('outliers for', i,"-",len(result))
```

```python
def z_score(df, threshold = 7):
  zscore = np.abs((df -  df.mean())/ df.std())
  outliers = df[zscore > threshold]
  return outliers

for i in column_names:
  result = z_score(df[i])
  print('outliers for', i,"-",len(result))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split as tts
```

```python
mean = df['total_bedrooms'].mean()
mean
```

```python
df['total_bedrooms']=df['total_bedrooms'].fillna(mean)
```

```python
mean = df['ocean_proximity'].mode()[0]
mean
df['ocean_proximity']=df['ocean_proximity'].fillna(mean)
```

```python
mean = df['total_rooms'].median()
mean
df['total_rooms']=df['total_rooms'].fillna(mean)
```

```python
mean = df['households'].median()
mean
df['households']=df['households'].fillna(mean)
```

```python
mean = df['median_income'].median()
mean
df['median_income']=df['median_income'].fillna(mean)
```

```python
cat_cols = df.select_dtypes(include=['object']).columns.to_list()
```

```python
data2 = pd.get_dummies(df, columns=cat_cols, drop_first= True)
```

```python
x = data2.drop(columns = ['median_house_value', 'Unnamed: 11', 'Unnamed: 12'])
y = data2['median_house_value']
```

```python
x_train, x_test , y_train, y_test = tts(x, y, train_size = 0.8, random_state=12)
lr = LinearRegression()
```

```python
model = lr.fit(x_train, y_train)
model
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
r2 = r2_score(y_test, y_pred)
r2
```
